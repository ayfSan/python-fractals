```python
from ColabTurtle.Turtle import *

# Тест: рисуем квадрат
initializeTurtle()  # Запускаем черепашку
speed(10)  # Скорость (1-13, 13 — самая быстрая)
penup()  # Поднимаем перо
goto(100, 100)  # Идём в стартовую точку
pendown()  # Опускаем перо
color('blue')  # Цвет линии

for _ in range(4):  # Цикл для 4 сторон
    forward(200)  # Вперёд 200 пикселей
    right(90)  # Поворот направо на 90 градусов
```

```python
from ColabTurtle.Turtle import *

# Вариация 1: Фрактальное дерево (рекурсия: ветки делятся на меньшие)
initializeTurtle()  # Новая черепашка для дерева
speed(13)
bgcolor('lightgreen')  # Фон для атмосферы
penup()
goto(500, 500)  # Старт снизу
pendown()
left(0)  # Смотрим вверх
color('brown')

def draw_tree(length, depth):
    if depth == 0:  # Если глубина 0 — стоп (листья)
        color('green')
        forward(5)  # Маленький лист
        backward(5)
        color('brown')
        return
    forward(length)  # Рисуем ветку
    left(30)  # Левая ветвь
    draw_tree(length * 0.7, depth - 1)  # Рекурсия: меньшая ветка
    right(60)  # Правая ветвь
    draw_tree(length * 0.7, depth - 1)  # Ещё рекурсия
    left(30)  # Возврат
    backward(length)  # Назад к стволу

draw_tree(100, 8)  # Запуск: длина 100, глубина 8 (поменяй для экспериментов)
```
